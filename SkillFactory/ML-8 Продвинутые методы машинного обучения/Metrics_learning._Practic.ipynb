
## Работа с выбросами

Для работы с выбросами сделаем диапазоновые значения признаков, для этого будем использовать функцию

```python
def add_range(df, columns, range_list, onesided = True):
    for column in columns:
        name = column + '_range'
        df[name] = 0
        for i in range(len(range_list)-1):
            df.loc[(df[column] > range_list[i])&(df[column] <= range_list[i+1]), name] = i+1
            if onesided:
                df.loc[(df[column] >range_list[-1]), name] = len(range_list)
    df.drop(columns, axis = 1, inplace = True)
```

Эта функция add_range добавляет новые столбцы к датафрейму df, соответствующие диапазонам значений из указанных столбцов. Функция также удаляет исходные столбцы после создания новых.

```python
add_range(data,['BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT',
        'PAYMENTS', 'MINIMUM_PAYMENTS'], [0,500,1000,3000,5000,10000])
```

```python
add_range(data,['BALANCE_FREQUENCY', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY', 
         'CASH_ADVANCE_FREQUENCY', 'PRC_FULL_PAYMENT'] , np.linspace(0,1,11),onesided=False)
```

```python
add_range(data,['PURCHASES_TRX', 'CASH_ADVANCE_TRX'] ,[0,5,10,15,20,30,50,100] ,onesided=False)
```

## Кластеризация

Обучите модель `Kmeans`с количеством кластеров, полученном в задании 6.2, с помощью метода

```python
kmean= KMeans(k_clust)
kmean.fit(X) 
labels=kmean.labels_
```

Объединяем два DataFrame: исходный data и новый DataFrame, содержащий метки кластеров labels. Он выполняет это с использованием функции pd.concat() по оси столбцов (axis=1), создавая новый DataFrame clusters.

```python
#раскомментируйте код для получения информации о распределениях признаков для каждого кластера
for c in clusters:
     grid= sns.FacetGrid(clusters, col='cluster')
     grid.map(plt.hist, c)
```

Эти графики дают представление о распределении значений BALANCE_range внутри каждого кластера и помогают понять характеристики и поведение каждого кластера. Например, кластеры 0, 2 и 4 имеют значительное количество клиентов с балансом в диапазоне BALANCE_range 2, что может указывать на схожее поведение или характеристики этих клиентов. Кластеры 1, 3, 5, 6 и 7 имеют более разнообразное распределение балансом.

Визуализируем кластеры, предварительно уменьшив разменость с помощью PCA до 2-х


```python
# Здесь создается объект PCA, который будет искать две главные компоненты.
pca = PCA(2)
# Модель PCA обучается на данных X, что позволяет вычислить главные компоненты
pca.fit(X)
# Данные X преобразуются в новое пространство с помощью двух главных компонент, и результат сохраняется в X_PCA
X_PCA = pca.transform(X)
# Вывод формы массива
X_PCA.shape
```

### Задача 8.3 
C помощью атрибута `explained_variance_ratio_` выведите процент объясненной дисперсии по обоим из осей. В качестве ответа внесите максимальное значение, округленное до десятых.

```python
# Процент объясненной дисперсии
explained_variance = pca.explained_variance_ratio_
print(f"Процент объясненной дисперсии по осям: {explained_variance}")

# Максимальное значение, округленное до десятых
max_explained_variance = round(max(explained_variance), 1)
print(f"Максимальное значение объясненной дисперсии, округленное до десятых: {max_explained_variance}")
```

```python
x, y = X_PCA[:, 0], X_PCA[:, 1]

colors = {0: 'red',
          1: 'blue',
          2: 'green', 
          3: 'yellow', 
          4: 'orange',  
          5:'purple',
          6:'fuchsia',
          7:'pink'}

names = {i: f'cluster {i+1}' for i in range(k_clust)}
  
df = pd.DataFrame({'x': x, 'y':y, 'label':labels}) 
groups = df.groupby('label')

fig, ax = plt.subplots(figsize=(20, 13)) 

for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5,
            color=colors[name],label=names[name], mec='none')
    ax.set_aspect('auto')
    ax.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off')
    ax.tick_params(axis= 'y',which='both',left='off',top='off',labelleft='off')
    
ax.legend()
ax.set_title("Customers Segmentation based on their Credit Card usage bhaviour.")
plt.show()
```

Как мы видим, результат достаточно неочевидный, для более красивой визуализации вычислим косинусное расстояние от исходных данных и применим к ним PCA.

Напомним, что косинусное расстояние имеет вид: $$1 - \frac{u  v}{||u||||v||}$$
В модуле `sklearn.metrics.pairwise` реализовано вычитаемое в данной разнице в функции `cosine_similarity` предлагается реализовать функцию `cos_dist` косинусного расстояния на основе данной встроенной функции.

```python
def cos_dist(X):
     return 1 - cosine_similarity(X) # здесь реализация
```

```python
dist =cos_dist(X)

pca = PCA(2)
pca.fit(dist)
X_PCA = pca.transform(dist)
X_PCA.shape
```
