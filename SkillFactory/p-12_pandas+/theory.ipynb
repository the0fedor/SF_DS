```python
import pandas as pd
melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
```

```python
display(melb_df.info())
```

```python
# Преобразуйте столбец Date в формат datetime
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
display(melb_df['Date'])
```

```python
# выделите квартал (quarter) продажи объектов недвижимости
quarter = melb_df['Date'].dt.quarter.value_counts()
display(quarter)
```

```python
# список столбцов, которые мы не берём во внимание
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
# задаём максимальное число уникальных категорий
max_unique_count = 150
# цикл по именам столбцов
for col in melb_df.columns: 
    # проверяем условие
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: 
        # преобразуем тип столбца
        melb_df[col] = melb_df[col].astype('category') 
display(melb_df.info())
```

## 2. Сортировка данных в DataFrame

МЕТОД SORT_VALUES()

Для сортировки значений в DataFrame по значениям одного или нескольких столбцов используется метод sort_values().

by — имя или список имён столбцов, по значениям которых производится сортировка.

axis — ось, по которой производится сортировка (0 — строки, 1 — столбцы). По умолчанию сортировка производится по строкам.

ascending — сортировка по возрастанию (от меньшего к большему). По умолчанию параметр выставлен на True, для сортировки по убыванию (от большего к меньшему) необходимо выставить его на False.

ignore_index — создаются ли новые индексы в таблице. По умолчанию выставлен на False и сохраняет индексы изначальной таблицы.

inplace — производится ли замена исходной таблицы на отсортированную. По умолчанию параметр выставлен на False, то есть замены не производится. Чтобы переопределить исходную таблицу на отсортированную, необходимо выставить этот параметр на True.

```python
# Отсортируем таблицу по возрастанию цены объектов недвижимости (Price):
melb_df.sort_values(by='Price').head(10)
```

```python
# А теперь отсортируем таблицу по убыванию (от самой последней до самой первой) даты продажи объекта (Date). 
# Для этого выставим параметр ascending на False:
melb_df.sort_values(by='Date', ascending=False)
```

СОРТИРОВКА ПО ЗНАЧЕНИЯМ НЕСКОЛЬКИХ СТОЛБЦОВ

Так, например, отсортируем таблицу сначала по возрастанию расстояния от центра города (Distance), а затем — по возрастанию цены объекта (Price). Для того чтобы вывод был более наглядным, выделим каждую десятую строку из столбцов Distance и Price результирующей таблицы:

```python
melb_df.sort_values(by=['Distance', 'Price']).loc[::10, ['Distance', 'Price']]
```

Попробуйте поменять порядок следования столбцов в параметре by метода sort_values() и сравните результат. 

```python
melb_df.sort_values(by=['Distance', 'Price'], ascending=False).loc[::10, ['Distance', 'Price']]
```

КОМБИНИРОВАНИЕ СОРТИРОВКИ С ФИЛЬТРАЦИЕЙ

Найдём информацию о таунхаусах (Type), проданных компанией (SellerG) McGrath, у которых коэффициент соотношения площадей здания и участка (AreaRatio) меньше -0.8. Результат отсортируем по дате продажи (Date) в порядке возрастания, а после проведём сортировку по убыванию коэффициента соотношения площадей. Также обновим старые индексы на новые, установив параметр ignore_index на True. Для наглядности результата выберем из таблицы только столбцы Data и AreaRatio:

```python
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
# 2.2 Произведите сортировку столбца AreaRatio по убыванию
melb_df.sort_values(by='AreaRatio', ignore_index=True, ascending=False).loc[1558, ['BuildingArea']]
```

```python
# 2.3 Найдите таунхаусы (Type) с количеством жилых комнат (Rooms) больше 2
# создаем маски с фильтрами
mask1 = melb_df['Type'] == 'townhouse'
mask2 = melb_df['Rooms'] > 2
# отфильтрованные по маскам таблицы
melb_df[mask1 & mask2].sort_values(
    # сортируем по количестсву комнат а затем по средней площади комнат
    by=['Rooms', 'MeanRoomsSquare'],
    # количество комнат по возрасстанию, а площадь по убыванию
    ascending=[True, False],
    # заменяем индексы и обращаемся к 18 строке и столбцу Price
    ignore_index=True).loc[18, ['Price']]
```

## 3. Группировка данных в DataFrame

В библиотеке Pandas для группировки данных по одному или нескольким признакам можно использовать метод groupby().

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html

by — имя или список имён столбцов, по которым производится группировка.

axis — ось, по которой производится группировка (0 — строки, 1 — столбцы). По умолчанию группировка производится по строкам.

as_index — добавляется ли дополнительный индекс к таблице. По умолчанию установлен на True.


```python
# Применим агрегирующую функцию среднего к результату работы groupby()
melb_df.groupby(by='Type').mean(numeric_only=True)
```

```python
# давайте сравним средние цены на объекты в зависимости от их типа:
melb_df.groupby('Type')['Price'].mean()
```

```python
# выясним, какие регионы (Regionname) наиболее удалены от центра Мельбурна
melb_df.groupby('Regionname')['Distance'].min().sort_values(ascending=False)
```

Чтобы рассчитать несколько агрегирующих методов, можно воспользоваться методом agg(), который принимает список строк с названиями агрегаций.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.agg.html

```python
# построим таблицу для анализа продаж по месяцам
melb_df.groupby('MonthSale')['Price'].agg(
    ['count', 'mean', 'max']
).sort_values(by='count', ascending=False)
# найдём количество продаж, а также среднее и максимальное значения цен объектов недвижимости (Price)
```

Примечание. Если вам нужна полная информация обо всех основных статистических характеристиках внутри каждой группы, вы можете воспользоваться методом agg(), передав в качестве его параметра строку 'describe': melb_df.groupby('MonthSale')['Price'].agg('describe')

```python
# чтобы получить множество из агентств недвижимости, которые работают в каждом из регионов
melb_df.groupby('Regionname')['SellerG'].agg(
    		['nunique', set]
)
```

```python
melb_df.groupby('Rooms')['Price'].agg(['mean']).sort_values(by='mean', ascending=False)
# 3.1 Сгруппируйте данные по признаку количества комнат и найдите среднюю цену объектов недвижимости
```

```python
# 3.2 Какой регион имеет наименьшее стандартное отклонение по географической широте (Lattitude)?
melb_df.groupby('Regionname')['Lattitude'].agg(['describe'])
```

```python
# 3.3 Какая риелторская компания (SellerG) имеет наименьшую общую выручку за период с 1 мая по 1 сентября (включительно) 2017 года?
# обозначаем переменные с границами периода
date_start = pd.to_datetime("2017-05-01")
date_end = pd.to_datetime("2017-09-01")
# создаем переменную с фильтрованной по датам таблицей
filtred_data = melb_df[(melb_df['Date'] >= date_start) & (melb_df['Date'] <= date_end)]
# групируем данные по риелторам и опредлеям кто меньше всего заработал за период
filtred_data.groupby('SellerG')['Price'].sum().sort_values(ascending=True)
```
