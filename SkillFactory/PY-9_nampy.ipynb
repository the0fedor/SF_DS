```python
import numpy as np
array = [1, 2, 3, 4, 5]
def shuffle_seed(array):
    our_seed = np.random.randint(2**32)
    # (low=0, high=(2**32), size = 1, dtype = np.uint32)
    np.random.seed(our_seed)
    shuffled_array = np.random.permutation(array)

    return (shuffled_array, our_seed)


print(shuffle_seed(array))
```

```python
import numpy as np # импортируем библиотеку numpydef min_max_dist(*vectors):
# функция, которая принимает на вход неограниченное число векторов через запятую
    dists = list() # задаем пустой список
    for i in range(len(vectors)): # создаем в цикле список равный количеству векторов, это будут индексы-идентификаторы векторов
        for j in range(i + 1, len(vectors)): # в этом цикле тоже создаем индекс-идентификатор, но уже со сдвигом на 1, чтобы избежать повторений при расчете расстояний, так как расстояние между 1 вектором и 2 вектором = расстоянию между 2 и 1 и т.д.
            dists.append(np.linalg.norm(vectors[i] - vectors[j])) # добавляем в список  dists  вычисленную векторную норму (или расстояния между элементами в нашем наборе данных)
    return (min(dists), max(dists)) # возвращаем результат работы функции - минимальное и максимальное значение в списке dists в виде кортежа
```

```python
import numpy as np
vectors_list = list()
def any_normal(*vectors):
    for i in range(len(vectors)):
        for j in range(i + 1, len(vectors)):
            if np.dot(vectors[i], vectors[j]) == 0:
                return True
    return False

vec1 = np.array([2, 1])
vec2 = np.array([-1, 2])
vec3 = np.array([3,4])
print(any_normal(vec1, vec2, vec3))
```

```python
import numpy as np
def get_unique_loto(num):
    # создаем список чисел от 1 до 100
    sample = np.arange(1, 101)
    res = list()
    # цикл по каждому элементу в количестве num итераций
    for i in range(num):
        # в каждой итерации создаем массив 5на5 из случайно выбранных чисел из списка sample
        res.append(np.random.choice(sample, replace=False, size=(5, 5)))
    res = np.array(res)
    return res
```
