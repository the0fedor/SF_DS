```python
# 6.3 больше вопросов чем ответов
import pandas as pd
import os

def concat_user_files(path):
    data = pd.DataFrame()
    file_names = os.listdir(path)
    file_names.sort()
    for name in file_names:
        tmp_data = pd.read_csv(path + '/' + name)
        data = pd.concat([data, tmp_data], axis=0, ignore_index=True)
    data = data.drop_duplicates()
    return data
```

```python
import pandas as pd
# импортируем библиотеки и считываем таблицу
ratings_movies = pd.read_csv('data/ratings_movies.csv', sep=',')
```

Для решения задач нам понадобится выделить из признака title год выпуска фильма. Для этого напишем функцию get_year_release(arg).

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

Из модуля re нам понадобится только функция findall(), которая позволяет найти в строке все слова, удовлетворяющие шаблону. Мы находим в строке с названием фильма шаблон "(DDDD)" — четыре цифры, обёрнутых в скобки, что соответствует году выпуска фильма. Если такого шаблона не было найдено (год выпуска не указан), функция возвращает None (в таблице это будет помечено как пропуск).

Данная функция основана на регулярных выражениях и использует модуль re, с которым мы пока не работали.

→ Модуль re предназначен для поиска шаблонов в тексте и встроен в язык, поэтому не нуждается в установке.

```python
# 8.1 Создайте в таблице новый признак year_release, который соответствует году выпуска фильма.
# У скольких фильмов не указан год их выпуска?
ratings_movies['year_release'] = ratings_movies['title'].apply(get_year_release)
ratings_movies['year_release'].info()
display(100836-100818)
```

```python
# 8.2
# Какой фильм, выпущенный в 1999 году, получил наименьшую среднюю оценку зрителей?
# В качестве ответа запишите название этого фильма без указания года его выпуска
#  Произведите фильтрацию по условию «год выпуска»
movies_1999 = ratings_movies[ratings_movies['year_release'] == 1999]
# произведите группировку по названию фильма(title) и посчитайте среднее по столбцу rating для каждого фильма
grouped_movies = movies_1999.groupby(by='title').mean(numeric_only=True)
# Отсортируйте результат в порядке возрастания.
grouped_movies.sort_values(by='rating').head()
```

```python
# 8.3
# Какое сочетание жанров фильмов (genres), выпущенных в 2010 году, получило наименьшую среднюю оценку (rating)?
# Произведите фильтрацию по условию «год выпуска».
movies_2010 = ratings_movies[ratings_movies['year_release'] == 2010]
grouped_genres = movies_2010.groupby(by='genres').mean(numeric_only=True)
grouped_genres.sort_values(by='rating').head()
```

```python
# 8.4
# Какой пользователь (userId) посмотрел наибольшее количество различных (уникальных) комбинаций жанров (genres) фильмов? В качестве ответа запишите идентификатор этого пользователя.
# Сгруппируйте таблицу по пользователям и найдите число уникальных жанров для каждого пользователя.
users = ratings_movies.groupby('userId')['genres'].nunique()
# Отсортируйте результат по убыванию.
janr_diff = users.sort_values(ascending=False)
display(janr_diff.head())
```

```python
# 8.5
# Найдите пользователя, который выставил наименьшее количество оценок, но его средняя оценка фильмам наибольшая.
# Сгруппируйте таблицу по пользователям (userId)
ratings_movies.groupby('userId')['rating'].agg(
    # метод agg принимает список строк с названиями агригаций(встроенных функций) и применяет их
    ['count', 'mean']
# сортируем по новым столбцам
).sort_values(['count', 'mean'], ascending=[True, False])
```

```python
# 8.6
# Произведите фильтрацию по условию «год выпуска 2018»
movies_2018 = ratings_movies[ratings_movies['year_release'] == 2018]
# Сгруппируйте отфильтрованную таблицу по жанрам и найдите средний рейтинг,
grouped_movies = movies_2018.groupby('genres')['rating'].agg(['count', 'mean']
#а также количество выставленных оценок с помощью метода agg()
).sort_values(['count', 'mean'], ascending=[True, False])
# результат по столбцу со средним значением сортируем по убыванию
grouped_movies[grouped_movies['count'] > 10]['mean'].sort_values(ascending=False)
```

```python
# 8.7
ratings_movies['date'] = pd.to_datetime(ratings_movies['date'], dayfirst=True)
ratings_movies['year_rating'] = ratings_movies['date'].dt.year
# индексами которой являются годы выхода фильмов (year_rating), 
# столбцами — жанры (genres), 
# а в качестве значений выступают средние оценки фильмов.
pivot_raitings = ratings_movies.pivot_table(
    index='year_rating',
    columns='genres',
    values='rating',
    aggfunc='mean'
)
display(pivot_raitings.head())
```
