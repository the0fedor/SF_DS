```python
import pandas as pd
churn_data = pd.read_csv('data/churn.csv')
churn_data.head()
```

```python
# Задание 9.1 
import matplotlib.pyplot as plt
counts = churn_data['Exited'].value_counts()

counts.plot(kind='bar')
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Churn Data')
plt.show()
print (counts)
```
