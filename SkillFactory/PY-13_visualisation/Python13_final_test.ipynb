```python
import pandas as pd
churn_data = pd.read_csv('C:/Users/Vladimir/DS/Module2/data/churn.csv')
display(churn_data.head())
churn_data.shape
```

```python
churn_data.info()
```

```python
churn_data['Exited'] = churn_data['Exited'].astype('str')
churn_data.info()
```

```python
def replace_proc(Exited):
    for index in 'Exited':
        if index == '1':
            index ==  'Exited'
        else:
            index ==  'Loyal'
    return Exited
    
churn_data['Exited'] = churn_data['Exited'].apply(replace_proc)
display(churn_data.head())
```

```python

#import matplotlib.pyplot as plt
%matplotlib inline
churn_behaviour = churn_data['Exited'].value_counts()[:2]
display(churn_behaviour)

churn_behaviour.plot(
    kind='bar',
    grid=False,
    figsize=(3, 6),
    colormap='magma'
);
```
